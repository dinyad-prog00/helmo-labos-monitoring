```python
import pandas as pd
from pandas import read_csv
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
import os, glob
import warnings
warnings.filterwarnings('ignore')

```

```python
cpu=read_csv("../data/snmplabo/app/app-cpu-use.csv",error_bad_lines=False)
cpu['datetime']= mdates.num2date(mdates.datestr2num(cpu['datetime']))
cpu= cpu.groupby(cpu.columns.str.split('.').str[0],axis=1).sum()
cpuaff = cpu.sample(n=8,axis=1)
cpuaff["datetime"]=cpu["datetime"]
cpuaff.head(6)
```

```python
ram=read_csv("../data/snmplabo/app/app-mem-use.csv",error_bad_lines=False)
ram['datetime']= mdates.num2date(mdates.datestr2num(ram['datetime']))
ram= ram.groupby(ram.columns.str.split('.').str[0],axis=1).sum()
ramaff = ram.sample(n=8,axis=1)
ramaff["datetime"]=ram["datetime"]
ramaff.head(6)
```
